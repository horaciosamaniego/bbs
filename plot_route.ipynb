## Plotting

```python
aou = '7180'
route = fn.split('_')[1][:-4]
```

```python
#coordinates
routes = pd.read_csv('Routes.csv',encoding='latin-1')
routes['ruta'] = routes['CountryNum'].astype(str)  + routes['StateNum'].astype(str) + routes['Route'].astype(str)
xy = (routes[routes['ruta']==route]['Longitude'].values[0],routes[routes['ruta']==route]['Latitude'].values[0])
```

```python

# print(route)
x = "{0:0.2f}".format(xy[0])
y = "{0:0.2f}".format(xy[1])
fig, ax = plt.subplots(figsize=(12,6))
sns.set_theme(style="darkgrid")

sns.lineplot(x='Year',y=aou,data=d, marker='o', dashes=False,label=species[int(aou)])
ax.set_title('Route #: '+route+" ( "+x+","+y+" )")
ax.set_ylabel('Number of individuals')
```
